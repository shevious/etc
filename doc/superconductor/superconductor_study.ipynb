# BCS theory 개요

**Attractive force**

먼저 저온 금속에서 phonon을 매개로하는 electron-electron scattering이 attractive임을 보인다.

1. Harmonic oscillator를 이용한 creation / annihlation
2. Phonon의 모델링과 양자화
3. Electron-phonon interaction Hamiltonian의 구성
4. Perturbation theory를 이용한 electron-electron scattering의 계산
5. Attractive force에 대한 물리적 설명


**Cooper pair**

Momentum과 spin이 반대인 전자가 ground state(0도)에서 pair를 형성하여 bound state가 되는 것을 보인다.

**BCS theory**

1. Fermi surface 근처의 모든 전자들의 many particle wave function을 구성한다.
2. Mean Field Hamiltonian을 통하여 BCS wavefunction의 ground state energy를 구한다.
3. Bogoliubov transformation을 통하여 finite temperature에서 excited state의 energy를 구한다.
4. Ground state와 excited state와의 에너지 gap을 통하여 superconductivity를 설명한다.  
5. Energy gap과 critical temperature와의 상관관계를 구한다.

**Impurities**

1. 소량의 impurity가 superconductivity에 손상을 가하지 않는 이유에 대해서 설명한다.

**Appendix**

1. 3차원에서 일반적인 weak potential은 bound state를 형성할 수 없음을 보인다.

## Schrödinger equation

Schrodinger equation ($\hbar=1$):  
$$
i{\partial\over\partial t}|x,t\rangle=
\left[-{1\over2m}{\partial^2\over\partial x^2}+V\right]|x,t\rangle
$$
$x$와 $t$ 변수 분리를 이용해서 미분방정식 풀기:
$$
|x,t\rangle = e^{-iEt}|x\rangle
$$
Time independent Shrödinger equation:  
$$
\left[-{1\over2m}{\partial^2\over\partial x^2}+V\right]|x\rangle =
E|x\rangle
$$
$$
{i\over\sqrt{2m}}{\partial\over\partial x}\equiv p
$$
$$
{p^2\over 2m}+V\equiv H
$$
The time-independent Schrodinger equation:
$$
H|x\rangle =
E|x\rangle
$$

## Quantum Harmonic Oscillator

![](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcRAQtzmQ2B0LBknmDjxYqGFuxcYlmkScvtMbyBLiGZkmdaZJDaYEny5Q7q0if3aEcVnnZA&usqp=CAU)
$$
H = {p^2\over 2m}+{1\over2}m\omega^2x^2
$$
양자화:  
$$[x,p] = i$$
Creation and annhilation operator를 정의하면 다음과 같다.:  
Creation operator:  $$
a^\dagger \equiv \sqrt{m\omega\over2}\left(x-i{p\over m\omega}\right) $$
Annhilation operator:   
$$
a \equiv \sqrt{m\omega\over2}\left(x+i{p\over m\omega}\right)$$
$a,a^\dagger$에 대해 $x$를 표현:  
$$x ={1\over \sqrt{2m\omega}}(a+a^\dagger)$$
$a,a^\dagger$의  commutation relation은 $[x,p] = i$로 부터 다음과 같아진다.  
$$[a,a^\dagger]=1$$
위의 식은 boson의 양자화와 동일하다.

Hamiltonian을 $a,a^\dagger$에 대해 재정의:
$$H=\omega\left(a^\dagger a+{1\over2}\right)$$
Schrodinger 방정식을 풀어서 engergy가 $E\equiv\omega (n+1/2)$이라고 하자. ($n$은 임의의 실수이고, $\omega$와 $1/2$는 편의를 위해 도입.) 이때의 wavefunction을 $|n\rangle$이라고 하면,
$$
H|n\rangle = \omega \left(n+{1\over2}\right)|n\rangle\\
=\omega(a^\dagger a+1/2)|n\rangle.
$$
따라서,
$$
a^\dagger a|n\rangle = n|n\rangle .
$$
이 때에 $a|n\rangle$을 $H$에 적용하고 $[a,a^\dagger]=1$를 이용하면,
$$
\begin{align*}
Ha|n\rangle &= \omega a^\dagger a a|n\rangle+(\omega/2)a|n\rangle\\
&=\omega (aa^\dagger-1)a|n\rangle+(\omega/2)a|n\rangle\\
&=\omega\left(n-1+{1\over2}\right)a|n\rangle.
\end{align*}
$$
$a|n\rangle$ 역시  H의 해(Eigenvector)이고, 그 에너지(Eigenvalue)는 $\omega(n-1+1/2)$이다. (한단계 낮아 진다.)  따라서,
$$
a|n\rangle = c|n-1\rangle.
$$
절대값을 취해 보면,
$$
\begin{gather*}
\langle n|a^\dagger a|n\rangle = n=c^2\\
c = \sqrt n\\
\therefore a|n\rangle=\sqrt n|n-1\rangle
\end{gather*}
$$
마찬가지로 $a^\dagger$에 대해서는 다음과 같다.
$$
a^\dagger|n\rangle = \sqrt{n+1}|n+1\rangle
$$

만일 $n$이 실수 (예를 들면 1.5)라면, 에너지는 0.5, -0.5, -1.5, ... 등으로 $a$를 적용하면서 무한대로 낮아질 수 있다. 이것은 시스템의 에너지가 계속 낮아지면서 외부로 무한대의 에너지를 공급할 수 있으므로, 해가 될 수 없다. 따라서 $n$은 정수이고, $n=0$일 때, ground state가 된다. [^ref1]

## Electron-phonon interaction

전자와 lattice ion간의 interaction energy는 특정위치($\bf x$)와 ion의 거리에 따른 위치 에너지의 합계 $\sum_\mathbf RV(\mathbf x-\mathbf X)$에 밀도를 곱하고 공간에 대해 적분한 것이다.

$\rho$: density of electron  
$V$: potential with lattice atomic system  
$\bf R$: Lattice points ($=ja$)  
$\bf r_R$: displacement from lattice points  
$\bf X = R+r_R$: absolute position of atoms
Hamiltonian:  
$$
H = \int d^3x{1\over N}\sum_{\bf R}\rho(\mathbf x)V({\bf x-X)}\tag{3.47}
$$

$\bf r_R$에 대해 1차항까지 테일러 전개하면 다음과 같다.     
$$
H = \int d^3x{1\over N}\sum_{\bf R}\rho(\mathbf x)
\left[
V({\bf x-R})+{\bf r_R}\cdot\nabla_{\bf R}V({\bf x-R})+\cdots)
\right]
$$


첫번째 항은 electron과 ion의 평형상태의 interaction이다. 이 interaction은 Hamiltonian의 free electron항에 이미 포함되어 있다. [^solyom] [^Marino2]  

두번째 항은 ion의 displacement에 의한 것으로 electron-phonon interaction에 해당한다.  

$$
\begin{align*}
H_{el-ph} &= {1\over\sqrt{N}}\sum_{\bf k,k',\sigma} c^\dagger_{\bf k\sigma}c_{\bf k'\sigma}
{1\over N}\sum_{\bf R}e^{i({\bf k'-k+q})\cdot\mathbf R_i}
\sqrt{\hbar\over 2m\omega_{\mathbf ks}}
[a_{\mathbf qs}+a^\dagger_{-\mathbf q,s}] \\
&=\sum_{\bf{k,q},s,\sigma}g_{\mathbf qs}c^\dagger_{\bf k+q,\sigma}c_{\bf k\sigma}[a_{\mathbf qs}+a^\dagger_{-\mathbf q,s}]
\\
&\equiv H_I
\end{align*}
$$

여기서 $g_{\mathbf ks}$는 interaction의 강도를 나타내는 것으로 다음과 같이 정의된다.  
$$
g_{\mathbf ks}\equiv \sqrt{\hbar\over 2mN\omega_{\mathbf qs}}
\int d^3x\ e^{-i{\bf q\cdot x}}\ \mathbf e_s\cdot\nabla_{\bf x}V(\mathbf x)
$$

Interaction term의 물리적 의미는 다음과 같다.
왼쪽 항은 $g_{\mathbf qs}c^\dagger_{\bf k+q,\sigma}c_{\bf k\sigma}a_{\mathbf qs}$로서 incoming electorn $c_{\bf k\sigma}$이 phonon을 흡수 $a_{\mathbf qs}$하여 outgoing electron $c^\dagger_{\bf k+q,\sigma}$이 방출되는 것이다.  오른쪽 항은 $\bf q \to -q$로 변경하여보면, $g_{-\mathbf q,s}c^\dagger_{\bf k-q,\sigma}c_{\bf k\sigma}a^\dagger_{\mathbf qs}$로서 incoming electorn $c_{\bf k\sigma}$이 phonon을 생성 $a^\dagger_{\mathbf qs}$하고 outgoing electron $c^\dagger_{\bf k-q,\sigma}$로 방출되는 것이다.

## Perturbation using Schrieffer-Wolff transformation

Canonical transformation을 이용한 이 purturbation theory는 degenerate perturbation theory의 한 종류이다.  

Hamiltonian이 다음과 같을때,
$$
H=H_0+\eta H_1
$$
Canonical transformation을 적용하면 다음과 같다.
$$
H' = e^{-\eta S}He^{\eta S}
$$
$S^\dagger = -S$ (skew hermition) 일 경우, 위의 변환은 unitary이다.  
  
이 변환된 $H'$에 대해 해를 구하면,
$$
\begin{gather*}
H'|\psi'\rangle = E|\psi'\rangle\\
He^{\eta S}|\psi'\rangle=Ee^{\eta S}|\psi'\rangle
\end{gather*}
$$
원래의 Hamiltonian과 같은 energy spectrum 얻을 수 있다.  
(ref: Sólyom)  

$H'$을 $\eta$에 대해 전개하면,
$$
\begin{aligned}
H' &= H+\eta[H,S]+{\eta^2\over 2}[[H,S],S]+\cdots\\
&=H_0+\eta(H_1+[H_0,S])+\eta^2[H_1,S]+{\eta^2\over2}[[H_0,S],S]
\end{aligned}
$$
1차 항을 0이되도록 하는 S의 조건은 다음과 같다.
$$
H_1+[H_0,S] = 0\tag{46}
$$
Transformed Hamiltonian:
$$
\begin{gather*}
H' = H_0+H_\text{eff}\\
H_\text{eff} = {\eta^2\over2}[H_1,S]
\end{gather*}
$$

**Fröhlich Hamiltonian**

$H_{e}$: free electron  
$H_{ph}$: free phonon  
$H_{e-ph}$: electron-phonon interaction
$$
\begin{gather*}
H = H_{e}+H_{ph}+H_{e-ph}\\
=\sum_{\bf k}\epsilon_{\bf k}c^\dagger_{\bf k}c_{\bf k}
+\sum_{\bf q}\omega_{\bf q}\left(a^\dagger_{\bf q}a_{\bf q}+{1\over2}\right)
+\sum_{\bf k,q}g_{\bf k,q}c^\dagger_{\bf k+q}c_{\bf k}(a_{\bf q}+a^\dagger_{\bf -q})
\end{gather*}
$$
  
$\eta H_1\to H_{e-ph}, \eta S\to S$로 놓고, 이제 $S$를 다음과 같이 추정(ansatz)한다.  
    
$$
S = \sum_{\bf k,q}g_{\bf k,q}c^\dagger_{\bf k+q}c_{\bf k}(x_{\bf kq}a_{\bf q}+y_{\bf kq}a^\dagger_{\bf -q})
$$

추정된 $S$의 $x_{\bf kq}, y_{\bf kq}$를 구하기 위해 commutation 값을 계산한다.  

$$
\begin{aligned}
\;[H_e, S] &= \sum_{\bf k,q}g_{\bf k,q}(\epsilon_{\bf k+q}-\epsilon_{\bf k})
c^\dagger_{\bf k+q}c_{\bf k}(x_{\bf kq}a_{\bf q}+y_{\bf kq}a^\dagger_{\bf -q})\\
\;[H_{ph}, S] &= \sum_{\bf k,q}g_{\bf k,q}
c^\dagger_{\bf k+q}c_{\bf k}
(-x_{\bf kq}\omega_{\bf q}a_{\bf q}+y_{\bf kq}\omega_{\bf -q}a^\dagger_{\bf -q})
\end{aligned}
$$
Using the relation $\omega_{\bf q}=\omega_{\bf -q}$,
$$
H_1+[H_0,S] = \sum_{\bf k,q}g_{\bf k,q}
c^\dagger_{\bf k+q}c_{\bf k}
\left[
(1+(\epsilon_{\bf k+q}-\epsilon_{\bf k}-\omega_{\bf q})x_{\bf kq})a_{\bf q}\right.\\
\left.
+(1+(\epsilon_{\bf k+q}-\epsilon_{\bf k}+\omega_{\bf q})
y_{\bf kq})
a^\dagger_{\bf -q}
\right]
$$
  
위의 식이 0이되어야 하는 조건으로부터  $S$를 구할 수 있다.
  
$$
\begin{aligned}
x_{\bf kq} = (\epsilon_{\bf k}-\epsilon_{\bf k+q}+\omega_{\bf q})^{-1}\\
y_{\bf kq} = (\epsilon_{\bf k}-\epsilon_{\bf k+q}-\omega_{\bf q})^{-1}
\end{aligned}
$$
이제 2nd perturbation term을 구한다.
$$
H_\text{eff} = {1\over2}\sum_{\bf k,k',q} g_{\bf k,q} g_{\bf k',-q}
( y_{\bf k',-q}-x_{\bf k',-q} )
c^\dagger_{\bf k+q}c_{\bf k}
c^\dagger_{\bf k'-q}c_{\bf k'}\\
\equiv \sum_{\bf k,k',q}  V_\text{eff}({\bf k, k', q})
c^\dagger_{\bf k+q}c_{\bf k}
c^\dagger_{\bf k'-q}c_{\bf k'}
$$

Phonon를 매개로하는 electron-electron scattering의 effective potential은 다음과 같다. [^Heid]

$$
V_{\text {eff}} =g_{\bf kq}g_{\bf k',-q}
{\omega_{\bf q}\over
(E_{\mathbf k'}-E_{\bf k'-q})^2-\omega^2_\mathbf q
}
$$

perturbation theory (quantum mechanics) 섹션 참고:

E = E_0+\langle\psi_0|H_I|\psi_0\rangle+\color{blue}{\langle\psi_0|H_I{1\over E_0-H_0}H_I|\psi_0\rangle}\\

$c^\dagger_{k+q}, c^\dagger_{k'-q}$는 두 전자의 outgoing(or creation)을 의미한다.

반경이 $k$인 momentum 구의 부피는 다음과 같다.  
$$
V_{sphere} = {4\pi\over3}k^3
$$

따라서 이 구에 존재할 수 있는 momemtum의 총 수는 다음과 같다.
$$
N = {V_{elementary}\over V_{sphere}}=\int_0^k g(k')dk'
$$

미분하여 k에 대한 density of state를 구하면 다음과 같다.

$$
g(k)dk = {V\over2\pi}k^2\ dk
$$

Debye는 flat dispersion $\omega = vk$를 가정하였다. $dk/d\omega=1/v$이므로,
$g(\omega)\ d\omega = g(k)\ dk$로부터, $\omega$에 대한 density of state를 구할 수 있다.

$$
g(\omega) = {V\over 2\pi v^3}\omega^2
$$

진동의 모드가 총 3가지 (한개의 longitudial, 두개의 transverse)이므로 density of state는 위 계산의 3배가 된다.

Lattice crystal 내에서 wave frequency에는 상한선이 존재한다. 이 상한선까지 density of state를 모두 적분하게되면 최대 자유도인 3N(crystal의 lattice point 수$\times$ 3개의 모드)가 된다.

$$
3N = 3\int_0^{\omega_\text{max}}g(w)dw ={V\over2\pi^2v^3}\omega_\text{max}^3
$$

이 최대 진동수를 Debye frequency라고 부른다. [^ref1]

$$
\omega_D \equiv \omega_\text{max} = v\left(6\pi^2{N\over V}\right)^{1\over3}
$$

## Attractive force의 기원

<div>
<img src="https://i.imgur.com/f7aFk6j.png" label="Annette" width="400"/>
</div>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**그림.** 두 전자가 scattering될 때의 전후 momentum의 가능한 범위.


두개의 전자가 phonon에 의에 매개될 때, scattering 전,후의 에너지는 Fermi 에너지 $\epsilon_F$로부터 $\pm kT$ 범위에 있다.  
온도가 Debye frequency의 범위보다 낮다면, interaction은 attractive이다.  
대략적으로 phonon의 frequency는 $\omega_D$라고 평균할 수 있다.    

이럴 경우 effective potential은 다음과 같다.
$$
V_{\text {eff}} =g_{\bf kq}g_{\bf k',-q}
{\omega_D\over
(E_{\mathbf k'}-E_{\bf k'-q})^2-\omega^2_D}
$$

이러한 attractive force의 기원은 다음과 같다.

첫째, 유한한 부피를 가진 metal내의 free electron은 상당히 큰 절대0도 Fermi energy를 갖는다.  
이 때문에, interaction하는 electron의 momentum은 이 부근에 모여있다.  
(평균 에너지=$\epsilon_F=10^4K$)  

둘째, phonon을 매개로하는 두 electron은 interaction 후에도 여전히 Fermi surface부근에 모여 있어야 한다.  
Fermi sea의 내부가 이미 다른 ground state들로 가득차 있기 때문이다.

셋째, scattering 전,후의 에너지 변화가 작을 경우(Debye 에너지보다 작을 경우), effective potential은 음수 즉, attractive force를 형성한다.  

넷째, 전자가 atom system을 지나갈 때의 time scale은 $\epsilon_F^{-1}$이다.  
atom이 활성화(excited states)된 후 지속시간은 $\omega_D^{-1}$이다.   
따라서 첫번째 전자가 지나간 후 한참 뒤에 두번째 전자가 이 excited state와 interact할 수 있다. 이 지연된(retarded) interaction은 Coulomb force에 의한 replusive interaction을 크게 감소시킨다.    

금속내 전자의 attractive force는 Fröhlich(1950, 1952), Bardeen과 Pines(1955) 그리고 Pines(1959) 등에 의하여 발표되었다.  

# Cooper pair

## Ground state of two electron

1. 두 전자의 momentum이 반대일 경우, 총 momentum의 합이 0이 되므로 두 입자의 총 에너지는 ground state에 가깝게 된다.  

2. 두 전자가 같은 spin을 가질 경우, exchange 에너지 때문에(exchange hole) interaction을 약화시킨다. 따라서 더 낮은 에너지를 갖기 위해서는 두 전자의 spin이 달라야 한다.[^MFYS]  [^Margin]  
    ![https://imgur.com/a/zGBsvRN](https://i.imgur.com/VsrPwCU.png)  
    **그림.** 두 전자의 위치 correlation 값.  
    전자가 움직일 때 같은 spin의 전자는 그 위치에 중첩될 수 없다.  
    전자의 주변에 같은 spin의 전자가 사라지는 exchange hole을 형성하게 된다.[^Annett]

Fermi level위에 두 전자의 wavefunction을 무게 중심 $\mathbf R_\text{cm}=({\bf r_1+r_2})/2$으로 분리하면 다음과 같다.

$$
\Psi(\mathbf r_1, \mathbf r_2)
= e^{-i\mathbf k_\text{cm}\cdot\mathbf R_\text{cm}}\varphi({\bf r_1-r_2})|\sigma_1\sigma_2\rangle
$$

두 전자의 총 momentum은 0일 경우 에너지가 최소가 된다. 따라서 ground state에서는 $\mathbf k_\text{cm}=0$ 이다. [^MFYS2]

Spin wavefunction은 spin singlet (S = 0)이거나,
$$|\sigma_1\sigma_2\rangle = {1\over\sqrt 2}(\left|\uparrow\downarrow\right\rangle - \left|\downarrow\uparrow\right\rangle)
$$
spin triplet(S=1)이다.
$$|\sigma_1\sigma_2\rangle =  \begin{cases}
 \left|\uparrow\uparrow\right\rangle \\
 {1\over\sqrt 2}(\left|\uparrow\downarrow\right\rangle + \left|\downarrow\uparrow\right\rangle) \\
\left|\downarrow\downarrow\right\rangle
\end{cases}
$$
BCS theory에서는 spin singlet을 가정한다.  
전자의 fermion 조건으로부터,
$$\Psi(\mathbf r_1, \mathbf r_2) = -\Psi(\mathbf r_2, \mathbf r_1)
$$

spin이 singlet(antisymmetry)일 경우, 위치는 symmetry가 되어서 더 낮은 ground state energy를 갖기 때문이다.  

$$
\varphi(\mathbf r_1-\mathbf r_2) = \varphi(\mathbf r_2-\mathbf r_1)
$$

이것을 Fourier 변환하면 다음과 같다.  

$$
\varphi(\mathbf r_1-\mathbf r_2) =
\sum_{\bf k}\varphi_\mathbf k e^{i\mathbf k(\mathbf r_1-\mathbf r_2)}
$$
$\varphi(\mathbf r)$가 even이므로 $\varphi_\mathbf k = \varphi_{-\mathbf k}$ 이다.   
Fermi sea는 모두 가득차 있는 것을 가정하므로, 이 때의 합계에는 $|\mathbf k| > k_F$의 조건이 있다.  
(수식에서는 이 조건을 편의상 생략한다.)  

전체의 pair wavefunction은 다음과 같다.  
$$
\Psi(\mathbf r_1, \mathbf r_2) =
\sum_{\bf k} \varphi_\mathbf k e^{i\mathbf k(\mathbf r_1-\mathbf r_2)} {1\over\sqrt 2}(\left|\uparrow\right\rangle_1\left|\downarrow\right\rangle_2 - \left|\downarrow\right\rangle_1\left|\uparrow\right\rangle_2)\tag{34.1.4}
$$

복잡한 1,2 index 대신에 second quantization form으로 표기하면 다음과 같다.[^solyom3] [^Park]
$$
|\Psi\rangle = \sum_{\bf k}\varphi_{\bf k}c^\dagger_{\bf k\uparrow}c^\dagger_{-\bf k\downarrow}|0\rangle
$$

두 electron의 $\mathbf k$는 Fermi surface근처에 있으므로, Effective potential을 다음과 같이 근사한다.
$$
V_{\text {eff}} =g_{\bf kq}g_{\bf k',-q}
{\omega_D\over
(E_{\mathbf k'}-E_{\bf k'-q})^2-\omega^2_D} \approx -g^2
$$

그러면 Fröhlich Hamiltonian에서 free phonon을 뺀 Hamiltonian은 다음과 같다. (spin sum 생략)  

$$
H = H_{e}+H_{e-ph}\\
=\sum_{\bf k}\epsilon_{\bf k}c^\dagger_{\bf k}c_{\bf k}
-g^2\sum_{\bf k,k',q}
c^\dagger_{\bf k+q}c_{\bf k}
c^\dagger_{\bf k'-q}c_{\bf k'}
$$

Wave function $|\Psi\rangle$는 $H|\Psi\rangle=E|\Psi\rangle$의 Scrhodinger 방정식을 만족한다. Schrodinger 방정식 왼쪽에 $c^\dagger_{\bf k}c^\dagger_{-\bf k}|0\rangle$를 곱하면  

$$
\langle0|c_{-\bf k}c_{\bf k} H|\Psi\rangle =
E\varphi_\mathbf k
=2\epsilon_\mathbf k\varphi_\mathbf k - g^2\sum_{\bf k'}\varphi_{\mathbf k'}
\tag{6.18}
$$

2는 $\epsilon_{-\mathbf k}, \epsilon_\mathbf k$의 두 항에서 $\epsilon_{-\mathbf k}=\epsilon_\mathbf k$를 이용하여 산출된다.
식 (6.18)을 $\varphi_\mathbf k$에 대해 풀면 다음과 같다.

$$
\varphi_\mathbf k = -\left(\sum_{\bf k'}\varphi_{\bf k'}\right){g^2\over E-2\epsilon_\mathbf k}
$$

양변을 $\bf k$에 대해 합계를 취하면 다음과 같다.  

$$
\sum_\mathbf k\varphi_\mathbf k = -\left(\sum_{\bf k'}\varphi_{\bf k'}\right)\sum_\mathbf k{g^2\over E-2\epsilon_\mathbf k}
$$
즉,
$$
1 = -\sum_\mathbf k{g^2\over E-2\epsilon_\mathbf k}.
$$

$\mathbf k$에 대한 합계를 density of states에 대한 적분으로 변환하면 다음과 같다.  

$$
1= -g^2g(\epsilon_F)\int_0^{\hbar\omega_D}d\epsilon{1\over E-2\epsilon}
$$

$\epsilon$을 0부터 적분하는 이유는, $\epsilon_\mathbf k$를 Fermi energy $\epsilon_F$와의 차이만큼만 계산하기 때문이다.   
적분을 수행하고 이것을 E에 대해 풀면 다음과 같다.  
$$
E = -2\hbar\omega_De^{-1/\Lambda}
$$

여기서 $\Lambda$는 electron-phonon coupling parameter라고 불리며 다음과 같이 정의된다.

$$
\Lambda \equiv g^2g(\epsilon_F)
$$

$\Lambda$는 보통 매우 작아서, 에너지는 Debye energy $\omega_D$보다 exponentially 작다.  

Debye energy가 100-300K인 것에 비해서 대부분의 metal superconductor에서 $T_c$ 1K 정도로 적어지는 원인이 된다.[^Annett2]  

흥미로운 것은 아무리 interaction이 적더라도 에너지는 음수이므로 bound state가 존재한다는 것이다. 이것을 **Cooper problem** 이라고 부른다.  

이것은 통상적으로는 볼 수 없는 이상한 현상이다. 왜냐하면 interaction이 없는 두 자유전자의 에너지는 $2\epsilon_\mathbf k + 2\epsilon_F$인데, 두 interacting 전자의 ground state 에너지는 $E + 2\epsilon_F < 2\epsilon_F $이기 때문이다. 아무리 interaction이 적더라도, $2\epsilon_\mathbf k$ 만큼의 에너지 저하가 일어났다고 볼 수 있다.  

특히 일반적인 3차원에서 아주 작은 attractive force로는 bound state를 형성할 수 없다.  (1차원과 2차원에서는 arbitarily small interaction에 대해서도 bound state가 존재한다.)   
Superconductor에서 weak interaction이 bound state를 만드는 이유는 전자의 energy가 fermi surface부근에만 존재하고 따라서 에너지 density가 $\rho(\epsilon)\to \rho(\epsilon_F)$로서 2차원과 마찬가지로 constant가 되기 때문이다. (See **Appendix - Bound states of weak potential**.)

# BCS Wave function

Cooper problem으로부터 BCS는 전체 Fermi surface 부근에서 모든 전자가 bound state를 형성하는 unstability를 가진다는 것을 깨달았다.  
Effective attractive interaction은 Fermi surface의 모든 전자가 Cooper pair bound state가 된다.  

다음 문제는 이러한 많은 전자를 기술하는 방법이었다.  
먼저,
$$
|\mathbf r_1\sigma_1,...,\mathbf r_N\sigma_N\rangle =
{1\over\sqrt{N!}}\sum_P(-1)^P
|\mathbf r_1\sigma_1,\mathbf r_2\sigma_2\rangle
|\mathbf r_3\sigma_3,\mathbf r_4\sigma_4\rangle\cdots
|\mathbf r_{N-1}\sigma_{N-1},\mathbf r_N\sigma_N\rangle
$$

위와 같은 방식으로 시도한 것은 복잡함 때문에 쉽지 않았다.  
그 대신에 Schrieffer는 **coherent state**로 기술하는 방법을 택했다.  

먼저 다음과 같은 Cooper pair creation operator를 정의한다.  

$$ P^\dagger_\mathbf k\equiv c^\dagger_{\bf k\uparrow}c^\dagger_{-\bf k\downarrow} $$

그리고 이 operator에 의한 many-body coherent state를 다음과 같이 정의하였다.  

$$ |\Psi_\text{BCS}\rangle = \prod_\mathbf kC_\mathbf k\exp\left(\alpha_\mathbf kP^\dagger_\mathbf k\right)|0\rangle $$

여기서 $\alpha_\mathbf k$는 complex variable로서 향후에 energy를 minimize하는 조건을 통하여 찾게 될 것이다.  

Pair operator는 Bose commutation 법칙을 만족하지 않지만,
$$ [P_{\bf k}, P^\dagger_{\bf k}] \neq 1 $$
동일한 operator끼리는 commute한다.  
$$[P^\dagger_{\bf k}, P^\dagger_{\bf k'}] =0 $$
또한, 동일 momentum 전자가 두 번 생성될 수 없으므로 ($c^\dagger_kc^\dagger_k=0$),
$$ (P^\dagger_\mathbf k)^2 = P^\dagger_\mathbf kP^\dagger_\mathbf k=0$$
이다. 이것을 이용하면 exponential이 간단해진다.  
$$
\prod_\mathbf k\exp \alpha_\mathbf kP^\dagger_\mathbf k
=\prod_\mathbf k\left(1+\alpha_\mathbf kP^\dagger_\mathbf k\right)
$$  
Wave function을 normalize 하면,
$$ 1 = C_\mathbf k^2\langle0|\left(1+\alpha^*_\mathbf kP_\mathbf k\right)
\left(1+\alpha_\mathbf kP^\dagger_\mathbf k\right)|0\rangle=
C_\mathbf k^2(1+|\alpha_\mathbf k|^2)
$$

$C_\mathbf k = \sqrt{1+|\alpha_\mathbf k|^2}$를 얻을 수 있다.
다음과 같이 변수를 정의하면,  
$$
u_\mathbf k = {1\over\sqrt{1+|\alpha_\mathbf k|^2}},\quad
v_\mathbf k = {\alpha_\mathbf k\over\sqrt{1+|\alpha_\mathbf k|^2}}\quad
\left(|u_\mathbf k|^2 + |v_\mathbf k|^2 = 1\right)
$$
Wave function은 다음과 같다.  
$$ |\Psi_\text{BCS}\rangle = \prod_\mathbf k
\left(u_\mathbf k+v_\mathbf kP^\dagger_\mathbf k\right)
|0\rangle $$

Shriefer는 일리노이 대학의 대학원생이던 1957년 겨울, 뉴욕 지하철을 타고가면서 위의 wave function에 대한 아이디어를 떠올렸다고 한다. [^Arovas2]

이제 $v_\mathbf k$의 의미를 알 수 있다.   
Spin up인 전자의 갯수를 구하면,
$$
\begin{align*}
\langle N_{\bf k\uparrow}\rangle &=
\langle \Psi_\text{BCS}|c^\dagger_{\bf k\uparrow}c_{\bf k\uparrow}
|\Psi_\text{BCS} \rangle\\
&= \langle0|(u^*_\mathbf k+v^*_\mathbf k c_{-\bf k\downarrow} c_{\bf k\uparrow})
c^\dagger_{\bf k\uparrow}c_{\bf k\uparrow}
(u_\mathbf k+v_\mathbf k c^\dagger_{-\bf k\downarrow} c^\dagger_{\bf k\uparrow}) |0\rangle\\
&=\langle0| v^*_\mathbf k c_{-\bf k\downarrow} v_\mathbf k c^\dagger_{-\bf k\downarrow}|0\rangle\\
&=|v_\mathbf k|^2.
\end{align*}
$$
마찬가지로, $\langle N_{\bf k\downarrow}\rangle = |v_\mathbf k|^2$이다.

따라서 총 전자의 수는 다음과 같다.  
$$
\langle N\rangle = 2 \sum_\mathbf k |v_\mathbf k|^2
$$

[^Arovas2]: Lecture Notes on Superconductivity - Arovas


# Mean Field Hamiltonian

Cooper pair와 관련되어 있는 부분만을 추출하면 effective Hamiltonian은 다음과 같다.

$$
H = \sum_{\bf k,\sigma}\epsilon_\mathbf k c^\dagger_{\bf k\sigma}c_{\bf k\sigma}
-g^2\sum_{\bf k', k}
c^\dagger_{\bf k'\uparrow}c^\dagger_{-\bf k'\downarrow}
c_{-\bf k\downarrow}c_{\bf k\uparrow}
$$

BCS wave function의 에너지는 다음을 계산하면 된다.

$$ E = \langle\Psi_\text{BCS}|H|\Psi_\text{BCS}\rangle $$

첫번째 free field 항은 다음과 같다.

$$\sum_{\bf k,\sigma}\epsilon_\mathbf k\langle c^\dagger_{\bf k\sigma}
c_{\bf k\sigma}\rangle = 2\sum_\mathbf k\epsilon_\mathbf k |v_\mathbf k|^2
$$

두번째 interaction 항은 다음과 같다.  
$$\begin{align*}
&\langle
c^\dagger_{\bf k\uparrow}c^\dagger_{-\bf k\downarrow}
c_{-\bf k'\downarrow}c_{\bf k'\uparrow} \rangle = \langle\Psi_\text{BCS}|P^\dagger_{\bf k}
P_{\bf k'}|\Psi_\text{BCS}\rangle\\
&=\langle0|\prod_{\bf k_2}(u^*_{\bf k_2}+v^*_{\bf k_2}P_{\bf k_2})
P^\dagger_{\bf k}
P_{\bf k'}
\prod_{\bf k_1}(u^*_{\bf k_1}+v^*_{\bf k_1}P_{\bf k_1}) |0\rangle \\
&=\langle0|\prod_{\bf k_2\neq k,k'}(u^*_{\bf k_2}+v^*_{\bf k_2}P_{\bf k_2})\\
&\quad\times (u^*_{\bf k}+v^*_{\bf k}P_{\bf k})
(u^*_{\bf k'}+v^*_{\bf k'}P_{\bf k'})
P^\dagger_{\bf k}
P_{\bf k'}(u_{\bf k}+v_{\bf k}P^\dagger_{\bf k})
(u_{\bf k'}+v_{\bf k'}P^\dagger_{\bf k'}) \\
&\quad\times \prod_{\bf k_1\neq k,k'}(u^*_{\bf k_1}+v^*_{\bf k_1}P_{\bf k_1}) |0\rangle\\
&=\langle0|\prod_{\bf k_2\neq k,k'}(u^*_{\bf k_2}+v^*_{\bf k_2}P_{\bf k_2})\\
&\quad\times (u^*_{\bf k}+v^*_{\bf k}P_{\bf k})
(u^*_{\bf k'}+v^*_{\bf k'}P_{\bf k'})
P^\dagger_{\bf k}
u_{\bf k}
v_{\bf k'} \\
&\quad\times \prod_{\bf k_1\neq k,k'}(u^*_{\bf k_1}+v^*_{\bf k_1}P_{\bf k_1}) |0\rangle\\
&= v^*_\mathbf k  u^*_{\bf k'} u_{\bf k}v_{\bf k'}= v^*_\mathbf k v_{\bf k'}u^*_{\bf k'} u_{\bf k}
\end{align*}
$$

총 에너지는 다음과 같다.
$$
E = 2\sum_\mathbf k\epsilon_\mathbf k |v_\mathbf k|^2  -
g^2\sum_{\bf k',k}v^*_\mathbf k v_{\bf k'} u^*_{\bf k'} u_{\bf k}\\
= 2\sum_\mathbf k\epsilon_\mathbf k\left( |v_\mathbf k|^2 -|u_\mathbf k|^2 -1\right)
-g^2\sum_{\bf k',k}v^*_\mathbf k v_{\bf k'} u^*_{\bf k'} u_{\bf k}
$$  

$u, v$를 독립 변수로 두고 다음의 두가지 constraint를 고려하여 E를 최소화하는 조건을 찾는다.

$$
\begin{gather*}
|u_\mathbf k|^2 + |v_\mathbf k|^2 = 1\\
N = \sum_\mathbf k \left( |v_\mathbf k|^2 -|u_\mathbf k|^2 -1\right)
\end{gather*}
$$
Lagrangian multiplier로서 $E_\mathbf k$와 $-\mu$를 도입하면,
$$
\begin{aligned}
{\partial E\over\partial u_\mathbf k} - \mu{\partial N\over\partial u_\mathbf k}+E_\mathbf k u^*_\mathbf k &= 0\\
{\partial E\over\partial v_\mathbf k} - \mu{\partial N\over\partial v_\mathbf k}+E_\mathbf k v^*_\mathbf k &= 0\text{ .}\end{aligned}
$$
다음과 같이 $\Delta$를 정의할 때,
$$
\Delta \equiv g^2\sum_{\bf k} u^*_{\bf k}v_{\bf k}\tag{6.55}
$$
다음의 연립방정식으로 부터,
$$
\begin{pmatrix}
\epsilon_{\bf k}-\mu & \Delta\\
\Delta^* & -(\epsilon_{\bf k}-\mu)
\end{pmatrix}
\begin{pmatrix}
u^*_{\bf k}\\
v^*_{\bf k}
\end{pmatrix} = E_{\bf k}
\begin{pmatrix}
u^*_{\bf k}\\
v^*_{\bf k}
\end{pmatrix} \tag{6.57}$$

BCS wave function의 에너지 eigenvalue를 얻을 수 있다.  

$$ E_{\bf k} = \pm\sqrt{(\epsilon_{\bf k}-\mu)^2 + |\Delta|^2} $$

이 eigenvalue와 eigenvector로 부터 다음의 값을 구할 수 있다.
$$
u^*_{\bf k}v_{\bf k} = {\Delta\over 2E_{\bf k}}\tag{6.61}
$$

식(6.55)로부터 **BCS gap equation**은 다음과 같다.

$$
\Delta = g^2\sum_{\bf k} {\Delta \over 2E_{\bf k}}
$$

합계를 적분으로 바꾸고 $\sum_{\bf k} \to g(\epsilon_F)\int d\epsilon$, $\epsilon \equiv \epsilon_{\bf k}-\mu$로 정의하면,[^solyom1] [^solyom2]

$$
\begin{gather*}
\Delta = \underbrace{g^2g(\epsilon_F)}_{\equiv\Lambda}
\int_{-\omega_D}^{\omega_D}d\epsilon {\Delta\over 2\sqrt{\epsilon^2+\Delta^2}}\\
{1\over\Lambda} = \sinh^{-1}\left(\omega_D\over\Delta\right)
\end{gather*}
$$

$\Delta \ll \omega_D$ 이므로, gap은 다음과 같다.

$$
|\Delta| \approx 2\omega_De^{-1/\Lambda}
$$

이 결과는 두 개의 전자로 이루어진 Cooper pair의 계산과 동일하다.

   Since there is no summation over the spin, only half of the density of states $\rho(\epsilon)$ should be taken.

# BCS engergy gap and quasiparticle states (Bogolon)

Hamiltonian의 interaction term은 다음과 같다.  
$$
c^\dagger_{\mathbf k\uparrow}c^\dagger_{-\mathbf k\downarrow}
c_{-\mathbf k'\downarrow}c_{\mathbf k'\downarrow} = P^\dagger_{\mathbf k} P_{\mathbf k'}
$$

이것을 평균에 대한 상대값으로 나눈다.  
$$
P^\dagger_{\mathbf k} = \langle P^\dagger_{\mathbf k}\rangle + \underbrace{P^\dagger_{\mathbf k}-\langle P^\dagger_{\mathbf k}\rangle}_{\equiv\delta}
$$

Interaction term에서 $\delta^2$을 무시하면 다음과 같다. [^Arovas]  

$$
P^\dagger_{\mathbf k} P_{\mathbf k'} = \underbrace{-\langle P^\dagger_{\mathbf k}\rangle\langle P_{\mathbf k'}\rangle}_\text{energy shift} +
\underbrace{\langle P_{\mathbf k'}\rangle P^\dagger_{\mathbf k}  +
\langle P^\dagger_{\mathbf k}\rangle P_{\mathbf k'} }_\text{keep this} +
\underbrace{\delta^2}_\text{drop this!}
$$

첫번째 energy shift 항은 interaction이 없는 에너지의 증감을 나타낸다.  
두번째 항을 이용한 interaction Hamiltonian은 다음과 같다.  

$$
-g^2\sum_{\bf k, k'}\left(\langle  c_{-\mathbf k'\downarrow}c_{\mathbf k'\uparrow}\rangle c^\dagger_{\mathbf k\uparrow}c^\dagger_{-\mathbf k\downarrow} + \langle c^\dagger_{\mathbf k\uparrow}c^\dagger_{-\mathbf k\downarrow}\rangle c_{-\mathbf k'\downarrow}c_{\mathbf k'\uparrow}
\right)  =
-\sum_{\bf k}\left(\Delta^*c^\dagger_{\mathbf k\uparrow}c^\dagger_{-\mathbf k\downarrow} + \Delta c_{-\mathbf k'\downarrow}c_{\mathbf k'\uparrow}\right)
$$

위의 식을 이용한 reduced Hamiltonian은 다음과 같다.

$$
\begin{aligned}
H &= \sum_{\mathbf k,\sigma} \epsilon_\mathbf k c^\dagger_{\mathbf k, \sigma}c_{\mathbf k, \sigma} -\sum_{\bf k}\left(\Delta^*c^\dagger_{\mathbf k\uparrow}c^\dagger_{-\mathbf k\downarrow} + \Delta c_{-\mathbf k'\downarrow}c_{\mathbf k'\uparrow}\right) +\underbrace{\text{ constant }}_\text{neglect energy shift} -\mu N \\
&=\sum_\mathbf k
\begin{pmatrix}
 c^\dagger_{\bf k\uparrow} &  c_{-\bf k\downarrow}
\end{pmatrix}
\begin{pmatrix}
\epsilon_\mathbf k -\mu& -\Delta \\
-\Delta^* & -(\epsilon_\mathbf k -\mu)
\end{pmatrix}
\begin{pmatrix}
 c_{\bf k\uparrow} \\
 c^\dagger_{-\bf k\downarrow}
\end{pmatrix}
\end{aligned}
$$

마지막에 추가된  $-\mu N=-\mu\sum_{\bf k,\sigma}c^\dagger_{\bf k\sigma}c_{\bf k\sigma}$은 interaction term이 총 pair number를 보존하지 않기 때문에 추가된 chemical potential 항이다. [^solyom]  
원래 Hamiltonian이 $c^\dagger, c$에 대한 4차 interaction인데 비해서 reduced Hamiltonian은 2차항만을 포함하고 있다.

이것은 Bogoliubov transformaion을 통하여 대각화 할 수 있다.  

$$
H = \sum_\mathbf k
\begin{pmatrix}
 b^\dagger_{\bf k\uparrow} &  b_{-\bf k\downarrow}
\end{pmatrix}
\begin{pmatrix}
E_\mathbf k & 0 \\
0 & -E_\mathbf k
\end{pmatrix}
\begin{pmatrix}
 b_{\bf k\uparrow} \\
 b^\dagger_{-\bf k\downarrow}
\end{pmatrix}\tag{6.79}
$$

이 Eigenvector를 bogolon이라고 부르며 다음과 같이 정의 된다.  

$$
\begin{aligned}
b_{\bf k\uparrow} = u_{\bf k}c_{\bf k\uparrow}-v_{\bf k}c^\dagger_{\bf -k\downarrow}\\
b^\dagger_{-\bf k\uparrow} = v^*_{\bf k}c_{\bf k\uparrow}+u^*_{\bf k}c^\dagger_{\bf -k\downarrow}
\end{aligned}
$$

식 (6.57}로부터 $|u_{\bf k}|^2$과 $|v_{\bf k}|^2$을 얻어서 그래프로 그려보면 다음과 같다.

![https://imgur.com/8wljkD7](https://i.imgur.com/8wljkD7.png)

$\mathbf k$가 작을 때에는 $v\approx 1$이므로 $v$는 electron과 같고, 반대로 $u$는 hole과 같다. Bogolon은 $u$와 $v$가 mix되어 있으므로 electron과 hole이 mix된 quasiparticle state이고 fermion의 anti-commutation relation을 따른다. 또한,
$$
\begin{aligned}
b_{\bf k\uparrow}|\Psi_\text{BCS}\rangle = 0\\
b_{-\bf k\downarrow}|\Psi_\text{BCS}\rangle = 0
\end{aligned}
$$
이므로 BCS ground state에서 bogolon은 존재하지 않는다.

Bogolon의 생성연사자를 한쌍의 cooper pair에 적용해보면 다음과 같다. [^MFYS]
$$
b^\dagger_{\bf k\uparrow}\left(u_{\bf k}+v_{\bf k}c^\dagger_{\bf k\uparrow}c^\dagger_{-\bf k\downarrow}\right)|0\rangle = c^\dagger_{\bf k\uparrow}|0\rangle\tag{6.27}
$$

즉, Bogolon을 생성하면 Cooper pair가 파괴된다는 것을 의미한다.

$$
\require{mhchem}
0 \xleftarrow{b_{\bf k\uparrow}}
\begin{Bmatrix}
\text{Cooper pair}\\
\left(u_{\bf k}+v_{\bf k}c^\dagger_{\bf k\uparrow}c^\dagger_{-\bf k\downarrow}\right)|0\rangle
\end{Bmatrix}
\xrightleftharpoons[b_{\bf k\uparrow}]{b^\dagger_{\bf k\uparrow}}
\begin{Bmatrix}
\text{Broken pair}\\
c^\dagger_{\bf k\uparrow}|0\rangle
\end{Bmatrix}
\xrightarrow{b^\dagger_{\bf k\uparrow}} 0
$$

온도가 높아질 경우 bogolon은 Fermi-Dirac distribution인 $f(E) = 1/(e^{\beta E}+1)$을 따를 것이다.

$$
\begin{aligned}
\langle b^\dagger_{\bf k\uparrow}b_{\bf k\uparrow}\rangle &= f(E_{\bf k})\\
\langle b_{-\bf k\downarrow}b^\dagger_{-\bf k\downarrow}\rangle &= \langle 1 - b^\dagger_{-\bf k\downarrow}b_{-\bf k\downarrow}\rangle = 1-f(E_{\bf k})
\end{aligned}
$$

이것으로 finite temperature에서의 BCS gap parameter를 구할 수 있다.  
$$
\begin{aligned}
\Delta &= g^2\sum_{\bf k}\langle c_{-\bf k\downarrow}c_{\bf k\uparrow}\rangle\\
&= g^2\sum_{\bf k}u_{\bf k}v^*_{\bf k}\left(1-2f(E)\right)
\end{aligned}\tag{6.88}
$$

# Energy gap means superconductivity

식 (6.79)로부터 bogolon의 energy eigenvalue는 다음과 같다. $$
E_{\mathbf k} = \pm\sqrt{(\epsilon_\mathbf k-\mu)^2 + |\Delta|^2}
$$

이것을 그래프로 그려보면 다음과 같다.  

![https://imgur.com/RdN9tUT](https://i.imgur.com/RdN9tUT.png])  

$\Delta = 0$ 인 Normal state에서는 electron을 추가하는 excited energy는 $\epsilon_{\bf k}$이고 electron을 제거하는 (hole을 추가하는) energy는 $-\epsilon_{\bf k}$이다.  
Superconducting state에서는 bogolon을 추가하는데에 $E_{\bf k}$만큼 에너지가 소요되고 이것은 에너지 갭 $\Delta$ 보다 커야 한다.

이것이 superconductivity를 만드는 열쇠가 된다. Excited state가 되면 momenum이 변경되게 되고 이것은 전류의 superflow를 불가능하게한다. 에너지 gap이 존재할 경우 low 에너지(저온) 상태에서 excited state로의 전이를 방지한다. [^Lancaster]  

보통의 metal에서는 작은 양의 에너지를 물질에 전달함으로서 electron scattering이 일어나고 전류가 감소한다.
Superconductor에서는 작은(infinitesimal) 양의 에너지를 전달 할 수 있으며, 유일한 방법은 Cooper pair를 파괴하는 것이다. 이를 위해서는 일정한 양 이상의 에너지가 필요하다. 이것이 소실(dissipation)이 없는 흐름의 기원이고 저항이 사라지게 된다. [^MFYS] [^ref1] [^ref2] [^ref3]

전류가 없는 상태에서는 net momentum이 0이지만 전류가 형성될 경우, BCS의 ground state는 모두 동일한 drift velocity($\mathbf k_\text{cm}$)를 갖게 된다. [^Bardeen]

    How does the BCS theory explain superconductivity? The first thing to realize is that in superconductors, it is not that the electrons suddenly become immune from scattering off impurities and off each other (the effect that gives electrical resistance in a normal metal), it is simply that scattering does not affect conduction. The two electrons in a superconducting pair have equal and opposite momentum with respect to each other, although the pair as a whole can glide through a sample when a current flows. All the pairs are locked together and glide at the same speed. If a pair scatters with a phonon, then the two electrons may change their individual momenta but the gliding of the pair along the current direction continues completely unaffected. # Critical temperature

Gap equation (6.61)과 (6.88)로부터 $1-2f(E) = \tanh{\beta E\over2}$이므로,
$$
\Delta = g^2\sum_{\bf k} {\Delta\over\sqrt{\epsilon^2+|\Delta|^2}}\tanh\left({\sqrt{\epsilon^2+|\Delta|^2}\over 2kT}\right)
$$

Bardeen, Cooper, Shriefer는 다음의 **the BCS gap equation**을 산출하였다.

$$
1 = \Lambda\int_{-\hbar\omega_D}^{\hbar\omega_D} {d\epsilon\over2\sqrt{\epsilon^2+|\Delta|^2}}
\tanh\left({\sqrt{\epsilon^2+|\Delta|^2}\over 2kT}\right)
$$

위의 적분식을 계산하면 다음과 같다. [^solyom] [^ref1]

$$
{2\over\Lambda} = \log{\hbar\omega_D\over2kT}+\log{4e^\gamma\over\pi}-{7\zeta(3)\over8\pi^2}\left(\Delta\over kT\right)^2
$$  
Gap $\Delta=0$ 인 경우가 critical temperature에 해당한다.  

$$
\begin{aligned}
{2\over\Lambda} &= \log{\hbar\omega_D\over2kT_c}+\log{4e^\gamma\over\pi}
\end{aligned}
$$

위의 두 식으로부터 $T_c$와 energy gap과의 관계를 얻을 수 있다. [^zeta3]

$$
\Delta = \sqrt{8\pi^2\over7\zeta(3)}kT_c\sqrt{1-{T\over T_c}} = 3.06kT_c\sqrt{1-{T\over T_c}}
$$
즉,
$$
{\Delta(T)\over\Delta(0)} = \sqrt{1-{T\over T_c}}
$$

# Evidence of BCS theory

## Isotpope effect

Superconductor의 $T_c$는 interaction하는 phonon의 frequency에 비례한다.
Phonon의 frequency는 Debye frequency에 집중되어 있고 이것은 원자의 질량인 $M^{-1/2}$에 비례하여야 한다. (Harmonic oscillator와 동일하다.)    

$$
T_c\propto \omega_D \propto {1\over\sqrt M} = M^{-1/2} = M^{-\alpha}
$$

따라서 같은 금속에서 질량이 다른 동위원소의 $T_c$를 측정하고 $\alpha$가 0.5임을 확인함으로서 superconductivity가 electron-phonon interaction에 의한 것임을 검증할 수 있다.  
1950년 수은의 isotope effect 실험에서 $\alpha=0.5$에 일치하는 실험 결과를 얻었다.

하지만 절대0도에서는 상황이 다르다. Ground state에서 전자는 momentum을 갖지만 scattering될 수 없다. 왜냐하면 에너지를 얻을 수도, 더 낮은 에너지로 전이할 수도 없기 때문이다. 더 낮은 에너지는 이미 fermi sea에 의해서 다른 electron이 모두 차지하고 있기 때문이다.

특히 저온에서는 위의 입자모델과 달리 전자의 위치를 특정할 수 없게 된다. Free electron은 momentum eigen state로서 fermi sea를 채우고 있으며, 전자는 여러개의 이온에 걸쳐있게 된다. 기본적으로 free electron은 impurity나 disorder가 아닌 이상 결정의 격자 구조와 충돌하지 않고 이동할 수 있다.



## Coherece length of Cooper pair

Cooper pair는 momentum 공간에서의 bound state이다. Fermi energy 근처의 engergy gap $\Delta$내에서 $\bf k$와 $-\bf k$의 전자들이 wave packet을 형성한다. Momontum 공간에서 이 영역의 폭은 다음과 같다.  

$$
\hbar\Delta k\approx {\Delta\over v_\text F}
$$

따라서 wave packet의 길이는 다음과 같다.
$$
\xi_0\approx{\hbar v_\text{F}\over\Delta}
$$

여기서 길이 $\xi_0$를 **coherence length**라고 부른다. 비슷한 결과를 (34.1.4)의 wave function을 이용하여 $\langle r^2\rangle$을 계산하여도 얻을 수 있다. [^solyom]

이 coherence length는 BCS theory의 cooper pair bound state의 물리적 크기를 나타낸다는 의미를 갖는다. [^Annett]  

>    cooper pair distance:  ~수백nm  
>    atom distance of metal: ~수A    

$\Delta$가 Fermi 에너지보다 훨씬 작기 때문에, coherence length는 $10^{-6} - 10^{-7}$ 정도로 크다. 이것은 atomic dimension보다 100배~1000배 크다. [^solyom]    

![](https://i.imgur.com/sub4FPJ.png)  

부피를 기준으로 할경우 1개의 impurity의 영향은 약 $10^{-6}\sim10^{-9}$ 정도이다. 이것은 작은 perturbation으로 이 때의 에너지가 energy gap $2\Delta$를 넘어서지 않는다면 scattering은 일어날 수 없고 cooper pair를 파괴하지도 않는다.  

Fermi energy가 $\epsilon_F\sim1\text{ eV}$ 이고 energy gap이 $\Delta\sim 1 \text{meV}$인 저온의 conventional superconductor에서 cooper pair의 effective volume내에는 약 $\sim10^6$ 개의 cooper pair가 중첩되어 있다. [^Kadin2]  

## Impurity와의 elastic scattering에 따른 critical current density

두 개의 전자가 Cooper pair를 이룰 때 finite한 static current가 흐를 경우의 center of mass momentum을 $\bf q$라고 하면, 각각의 전자는 $\mathbf k+\mathbf q/2$와 $-\mathbf k+\mathbf q/2$의 momentum을 갖는다.  
그렇다면 전자의 속도는 $-\hbar\mathbf q/2m_e$이다. 따라서 current density는 다음과 같다.  

$$
j = -en_e{\hbar q\over 2m_e}
$$

Impurity에 의해 Cooper pair가 파괴 되려면 BCS enenergy gap $2\Delta$보다 커야한다. $(\mathbf k+\mathbf q/2,\mathbf k-\mathbf q/2)$ pair 중 앞의 것이 충돌하여 momentum이 $-\mathbf k+\mathbf q/2$가 되었다고 하자.  충돌 전후 전자의 에너지 차이는 gap을 넘어서야 한다.  

$$
{\hbar^2(\mathbf k+\mathbf q/2)^2\over 2m_e} - {\hbar^2(-\mathbf k+\mathbf q/2)^2\over 2m_e}\ge 2\Delta
$$

이것으로 전자의 최소 drift 속도를 구할 수 있다.

$$
q{\hbar^2k_\text F\over m_e}\ge 2\Delta
$$

이 조건을 만족시키는 critical current density는 다음과 같다.

$$
j_c = -{en_e\Delta\over\hbar k_\text F}
$$

Gap이 1meV라고 할 때 critical current density는 $10^7 \text{A/cm}^2$ 이다. (Superconductor의 바깥쪽의 magnetic field의 침투거리를 고려하지 않은 것으로, 실험으로 측정된 것 보다는 수 order의 차이가 난다.)

즉, superconductor의 전류밀도가 엄청나게 높지 않은한, impurity는 Cooper pair를 파괴할 수 없고 따라서 superconductivity가 유지된다.

## Bound states of weak potential

두 개의 particle이 약한 interaction을 하는 것으로 가정한다.  
Interaction은 반경이 $R$이하일 때 위치에너지가 $-V_0$인 finite potential well이다.  

Scrodinger 방정식은 다음과 같다. (reduced mass  $\mu={m\over2}$)

$$
\left({1\over m}{\partial^2\over\partial x^2}+E\right)\psi(x) = V(x)\psi(x)
$$
Fourier transform을 하면,
$$
\begin{gather*}
\left({1\over m}{\partial^2\over\partial x^2}+E\right)\int{dp\over2\pi}e^{-ipx}\psi(p) = \int{dp\over2\pi}e^{-ipx}\int dx'e^{ipx}V(x')\psi(x')\\
\left(-{p^2\over m}+E\right)\psi(p) = \int {dp'\over2\pi}V(p-p')\psi(p')\\
\psi(p) = -mV_0{1\over p^2+k^2}\int {dp'\over2\pi}\psi(p')\\
-{k^2\over m}\equiv E
\end{gather*}
$$
양변을 p에 대해 적분하면 (self consistancy),
$$
-{1\over V_0} = \int {dp\over2\pi}{m\over p^2+k^2}\approx \int_{p<{1\over R}} {dp\over2\pi}{m\over p^2+k^2}
$$
일반적으로 d차원에서는,
$$
\begin{gather*}
-{1\over V_0} = \int_{p<{1\over R}} {dp^d\over(2\pi)^d}{m\over p^2+k^2}
= \int_{\epsilon < E_R}d\epsilon{\rho_d(\epsilon)/\Omega\over 2\epsilon+|E|}\\
E_R\equiv {\hbar^2\over mR^2}
\end{gather*}
$$

여기서 weak potential $V_0\sim VR^d$이고, $V_0$가 volume의 차원을 가지므로 $\Omega$는 system의 volume이다.

2차원에 대해 계산하면 다음과 같다.  
$$
\begin{gather*}
d^2p = 2\pi p\ dp \\
\epsilon = {p^2\over 2m} \to p\ dp = m\ d\epsilon \\
{d^2 p\over(2\pi)^2} = {m \over2\pi}\ d\epsilon \equiv {\rho_{2D}(\epsilon)\over\Omega}d\epsilon \ \to\  {\rho_{2D}(\epsilon)\over m\Omega} = {1\over2\pi} =\text{constant}
\end{gather*}
$$

3차원에 대해 계산하면 다음과 같다.

$$
\begin{gather*}
{d^3p\over(2\pi)^3} = {1\over (2\pi)^3}4\pi p^2\ dp = {\sqrt{2m\epsilon}\over2\pi^2}m\ d\epsilon\equiv{\rho_{3D}(\epsilon)\over\Omega}d\epsilon \\
{\rho_{3D}(\epsilon)\over m\Omega} = {\sqrt{2m\epsilon}\over2\pi^2}
\end{gather*}
$$

적분을 통하여 $-1/V_0$의 값을 구하면 다음과 같다. [^Ketterle]

2차원:  
$$
-{1\over V_0}= \text{const.} \log {2E_R + |E|\over|E|}
$$
3차원:  
$$
-{1\over V_0}=\text{const.}\left( \sqrt{2E_R} - {\pi\over2}\sqrt{|E|}\right)
$$

$|V_0|=-V_0\to 0$일 경우 왼쪽항은 $+\infty$로 발산한다.  
1차원과 2차원에서는 $-V_0=|V_0|$가 작더라도 그것을 만족시키는 $|E|$의 해가 존재하지만, 3차원에서는 존재하지 않는다.  
따라서 3차원에서 weak potential은 bound state를 형성할 수 없다.

## Perturbation theory (quantum mechanics)

Total Hamiltonian = Free particle Hamiltonian  + Interaction Hamiltonian
$$
H =H_0+H_I
$$
Without interaction (or free particle term):  
$$
H_0|0\rangle = E_0|0\rangle
$$
Turn on the interaction:
$$
(H_0+H_I)(|0\rangle+|0^{(1)}\rangle=(E_0+E_0^{(1)})(|0\rangle+|0^{(1)}\rangle)
$$
$H_I, |0^{(1)}\rangle,E_0^{(1)}\ {\sim}\ \epsilon$: collecting 1st order term:  
$$
(H_I-E^{(1)}_0)|0\rangle = (E_0-H_0)|0^{(1)}\rangle
$$
Appling $\langle 0|\times$  :
$$
E^{(1)}_0 = \langle0|H_I|0\rangle
$$
Appling $\langle n|\times$  :
$$
|0^{(1)}\rangle =\sum_{n\neq0}{\langle n|H_I|0\rangle\over E_0-E_n}|n\rangle
$$
Adding 2nd order terms:  
$$
(H_0+H_I)(|0\rangle+|0^{(1)}+|0^{(2)}\rangle=  (E_0+E_0^{(1)}+E_0^{(2)})(|0\rangle+|0^{(1)}+|0^{(2)}\rangle
$$
Collecting 2nd order terms:  
$$
(H_I-E_0^{(1)}) |0^{(1)}\rangle=
E_0^{(2)}|0\rangle+
(E_0-H_0)|0^{(2)}\rangle
$$
Appling $\langle 0|\times$  :
$$
E_0^{(2)} = \langle0|H_I|0^{(1)}\rangle=
\sum_{n\neq0}
{\langle 0|H_I|n\rangle\langle n|H_I|0\rangle
\over E_0-E_n}\\
=\langle 0|H_I{1-P\over E_0-H_0}H_I|0\rangle
$$
where $P$ is the projectio operator to the ground state.
$$
P = |0\rangle\langle0|\\
1-P = \sum_{n\neq0}|n\rangle\langle n|
$$
By repeating this process:
$$
E_0^{(3)}
=\langle 0|H_I{1-P\over E_0-H_0}H_I{1-P\over E_0-H_0}H_I|0\rangle
$$

### Marino  

The effective total electron-phonon Hamiltonian, including the phonon-induced
effective electronic interaction, is therefore given by
$$
H_T = H_0+\sum_{\bf k, k',q}\sum_{\sigma,\sigma'}
V({\bf k, q}) c^\dagger_{\bf k'+q,\sigma'}c^\dagger_{\bf k-q,\sigma}c_{\bf k,\sigma}c_{\bf k',\sigma'}
$$
$$
V({\bf k, q}) = |W_{\bf q}|^2 \left[\hbar\omega\over(\epsilon_{\bf k}-\epsilon_{\bf k-q})^2-(\hbar\omega)^2\right]\tag{3.62}
$$

We will see that this Hamiltonian plays a central role in the explanation of superconductivity and forms the basis of the BCS theory, which describes this phenomenon. Curiously, however, it also describes the interaction that is the main cause of a nonzero resistance in the charge transport in solids.  
With the above study about the electron-phonon interaction, we conclude the description of the main electronic interactions in a crystal. In the next chapter we study important effects of these interactions.


A detailed study of the electron-phonon interaction expresses the effective electron-electron inter- action, mediated by one-phonon exchange. This expression is remarkable for two reasons. First, this effective interaction becomes negative (that is, attractive) when- ever the energy difference between the two electrons is less than the energy of the exchanged phonon. Secondly, if this energy difference is close to (but less than) the phonon energy, this attractive interaction becomes large and supersedes the Coulomb repulsion between the electrons.  

Remember, the phonon frequency is always less than the Debye frequency $\omega_D$.  It follows that when the electrons’ energy difference $\Delta\epsilon$ is larger than the Debye energy, $\Delta\epsilon >\hbar\omega_D$, the interaction (3.62) will never be attractive.  Conversely, for $\Delta\epsilon <\hbar\omega_D$, this will always be attractive for some range of the exchanged phonon energy. The former situation is the one that is most fre- quently found in material systems: the electronic effective interaction is repulsive. This situation is clearly enhanced as we increase the temperature, thereby boosting the electrons’ energy differences. Nevertheless, when we lower the temperature in a metal, we reach a situation in which most of the electrons occupy the states below the Fermi surface. Then, a large number of electrons will be close to the Fermi sur- face, and therefore would have small energy differences. Hence, it is likely that a large number of electrons will be found inside a shell of width of the order of $\omega_D$ around the Fermi surface ($\omega_D \ll \epsilon_F$). For these electrons, the attractive interaction mediated by phonons (3.62) will be more important than the Coulomb repulsion, and two-electron bound states with opposite spins, called Cooper pairs, will consequently form. These Cooper pairs then form a condensate in which the order parameter has a constant modulus, namely, an incompressible fluid. As we saw above, this is the key condition for the existence of a persistent current. The Cooper pairs therefore are the carriers of the superconducting current.

The net effective electron-electron interaction, which takes into account both the Coulomb repulsion and the phonon-mediated attraction, is described by (3.62), with the interaction potential given by
$$
V_{\text{eff}}({\bf k, q}) =
\begin{cases}
 -\lambda & |\epsilon_{\bf k}-\epsilon_{\bf k-q}| < \hbar\omega_D \\
 \ \ \ 0 & |\epsilon_{\bf k}-\epsilon_{\bf k-q}| > \hbar\omega_D
\end{cases}\tag{4.82}
$$
where $\lambda > 0$ is an effective coupling constant.

We use quantum field theory methods to show that this effec- tive, phonon-induced attractive quartic electronic interaction indeed produces an incompressible fluid of Cooper pairs below a certain critical temperature $T_c$, which is determined as well. This fluid is described by a complex order parameter of constant modulus. As we have seen above, this automatically leads to a superconducting state containing a persistent current.

## Why attractive?

1. Fermi boundary effect

   저온에서 전자의 momentum은 Fermi level 근처에 있다. Interaction에 의해 한 전자가 에너지를 많이 얻으려면 다른 전자의 energy가 낮아져야 하는데, 이미 ground state들이 가득차 있으므로 그런 interaction이 불가능하다. 이런 특수한 조건에서 attractive force가 발생한다.  

2. Retarded force에 의한 Coulomb force의 감소

   전자가 방출한 phonon을 흡수할 때 이 속도는 상당히 느리다. 따라서 첫번째 전자가 방출한 힘을 상당히 나중에 두번째 전자가 받게된다. 이 때문에 coulomb force의 영향력이 적어진다.

Although Cooper pairing is a quantum effect, the reason for the pairing can be seen from a simplified classical explanation.[[2]](https://en.wikipedia.org/wiki/Cooper_pair#cite_note-Hyperphysics-2)[[3]](https://en.wikipedia.org/wiki/Cooper_pair#cite_note-3) An electron in a metal normally behaves as a free particle. The electron is repelled from other electrons due to their negative charge, but it also attracts the positive ions that make up the rigid lattice of the metal. This attraction distorts the ion lattice, moving the ions slightly toward the electron, increasing the positive charge density of the lattice in the vicinity. This positive charge can attract other electrons. At long distances, this attraction between electrons due to the displaced ions can overcome the electrons' repulsion due to their negative charge, and cause them to pair up. The rigorous quantum mechanical explanation shows that the effect is due to electron–phonon interactions, with the phonon being the collective motion of the positively-charged lattice.

At low temperatures for very pure metals the mean free path can be made as high as a few
